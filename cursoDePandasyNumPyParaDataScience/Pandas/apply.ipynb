```python
import pandas as pd
```

```python
books_df = pd.read_csv('bestsellers-with-categories.csv', sep=',', header=0, names=['Nombre', 'Autor', 'Calificación', 'Reviews', 'Precio', 'Año', 'Género'])
books_df.head(2)
```

```python
def two_times(value):
    return value * 2
```

```python
books_df['Calificación'].apply(two_times)
```

```python
books_df['Rating_2'] = books_df['Calificación'].apply(two_times)
books_df
```

```python
books_df['Rating_2'] = books_df['Calificación'].apply(lambda x : x * 3)
books_df
```

```python
books_df['Rating_2'] = books_df.apply(lambda x : x['Calificación'] * 2 if x['Género']=='Fiction' else x['Calificación'], axis=1)
books_df
```
