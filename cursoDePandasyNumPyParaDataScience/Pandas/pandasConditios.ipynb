```python
import pandas as pd
import numpy as np
```

```python
books_df = pd.read_csv('bestsellers-with-categories.csv', sep=',', header=0, names=['Nombre', 'Autor', 'Calificación', 'Reviews', 'Precio', 'Año', 'Género'])
books_df.head(2)
```

```python
mayor_a_2016 = books_df['Año'] > 2016
```

```python
books_df[mayor_a_2016]
```

```python
genero_ficcion = books_df['Género'] == 'Fiction'
```

```python
books_df[genero_ficcion & mayor_a_2016]
```

```python
books_df[~mayor_a_2016] #Use ~ to negate a condition
```

```python
books_df[~genero_ficcion & ~mayor_a_2016]
```
