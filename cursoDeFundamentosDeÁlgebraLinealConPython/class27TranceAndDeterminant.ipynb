```python
v1 = np.array([1,0])
v2 = np.array([0,1])
               
vectorPlot([v1,v2],["blue","red"],1)

plt.xlim(-1,2)
plt.ylim(-1,2)

plt.show()
```

```python
A = np.array([[2,0],[0,2]])

v1_tranf = A.dot(v1)
v2_tranf = A.dot(v2)

vectorPlot([v1_tranf,v2_tranf],["orange","green"],1)

plt.xlim(-1,3)
plt.ylim(-1,3)

plt.show()
```

```python
det_A = np.linalg.det(A)

print(det_A)
```

```python
B = np.array([[-2,0],[0,2]])

v1_tranf = B.dot(v1)
v2_tranf = B.dot(v2)

vectorPlot([v1_tranf,v2_tranf,v1,v2],["orange","green","red","blue"],1)

plt.xlim(-3,3)
plt.ylim(-1,3)

plt.show()
```
