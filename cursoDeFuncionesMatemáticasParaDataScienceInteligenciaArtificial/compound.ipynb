# Función compuesta

Conocidas las funciones $f$ y $g$, la composición de $f$ y $g$ está dada por:

$$f o g = (f o g)(x) = f(g(x))$$

```python
import numpy as np
import matplotlib.pyplot as plt

N = 1000

x = np.linspace(-10, 10, num=N)

def g(x):
    return x**2

def f(x):
    return np.sin(x)

f_o_g = f(g(x))
g_o_f = g(f(x))

fig, ax = plt.subplots(1, 2)
ax[0].plot(x, f_o_g)
ax[0].set_title("$f(g(x))$")

ax[1].plot(x, g_o_f)
ax[1].set_title("$g(f(x))$")
plt.show()
```
