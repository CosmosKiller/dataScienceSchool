```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
country = ['INDIA', 'JAPAN', 'MEXICO', 'COLOMBIA', 'GERMANY']
population =[1000, 800, 900, 1000, 300]

plt.style.use('seaborn-dark')
```

```python
plt.bar(country, population, width=0.5, color=['#5c58b6','#b457ce','#5994ce','#3a4e93'])
plt.xticks(np.arange(5), ('India','Japan','Mexico','Colombia', 'Germany'), rotation= 45)
plt.title('Population per country')
plt.xlabel('Country')
plt.ylabel('Population (Millions)')
plt.show()
```

```python
plt.barh(country, population, color=['#5c58b6','#b457ce','#5994ce','#3a4e93'])
plt.show()
```
