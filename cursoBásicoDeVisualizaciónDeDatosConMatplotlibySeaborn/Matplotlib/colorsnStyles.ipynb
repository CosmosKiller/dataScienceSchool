```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
```

```python
plt.style.available
```

```python
plt.style.use('dark_background')
```

## Matlab

```python
fig, ax = plt.subplots(figsize=(8, 8))

ax.plot(x, x+1, 'r--')
ax.plot(x, x+2, 'b-')
ax.plot(x, x+3, 'g-.')
ax.plot(x, x+4, 'yo:')
```

## Pyplot

```python
fig, ax = plt.subplots(figsize=(8, 8))

ax.plot(x, x+1, color = 'green', alpha= 0.5, linewidth= 5, linestyle= '-', marker= 'o', markersize= 10, markerfacecolor= 'white')
ax.plot(x, x+2, color = 'orange', alpha= 0.4,  linewidth= 2, linestyle= '--', marker= 'x')
ax.plot(x, x+3, color = 'violet', alpha= 0.6,  linewidth= 8, linestyle= 'dashed', marker= 'v')
ax.plot(x, x+4, color = 'red', alpha= 0.8,  linewidth= 4, linestyle= ':', marker= 'P')
```
