```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = np.sin(x)
```

```python
fig, axes = plt.subplots()
axes.plot(x, y, 'b')
```

```python
fig, axes = plt.subplots(ncols=2, nrows=1)
axes[0].plot(x, y, 'b')
axes[1].plot(y, x, 'r')
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, nrows=1)
ax1.plot(x, y, 'b')
ax2.plot(y, x, 'r')
```

```python
fig, axes = plt.subplots(2, 4) #plt.subplots(rows, columns)
axes[0, 0].plot(x, np.cos(x))
axes[0, 1].plot(x, np.sin(x), 'r')
axes[0, 2].plot(x, np.tan(x), 'y')
axes[0, 3].plot(x, np.cos(x)**2, 'g')
```

```python
fig, ((ax1, ax2, ax3, ax4), (ax5, ax6, ax7, ax8)) = plt.subplots(2, 4) 
ax1.plot(x, np.cos(x))
ax2.plot(x, np.sin(x), 'r')
ax3.plot(x, np.tan(x), 'y')
ax4.plot(x, np.cos(x)**2, 'g')
fig.tight_layout()
```

```python
fig, ((ax1, ax2, ax3, ax4), (ax5, ax6, ax7, ax8)) = plt.subplots(2, 4) 
ax1.plot(x, np.cos(x))
ax2.plot(x, np.sin(x), 'r')
ax3.plot(x, np.tan(x), 'y')
ax4.plot(x, np.cos(x)**2, 'g')
ax5.plot(x, x**2, 'orange')
ax6.plot(x, x**3, 'c')
ax7.plot(x, x/2, 'm')
ax8.plot(x, x**(1/2), 'k')
fig.tight_layout()
```
