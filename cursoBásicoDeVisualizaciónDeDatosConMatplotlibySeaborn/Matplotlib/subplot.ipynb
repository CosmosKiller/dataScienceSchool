```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = x**2
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'gH--')
plt.subplot(1, 2, 2)
plt.hist(y)
plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'gH--')
plt.plot(y, x, 'cx-')
plt.subplot(1, 2, 2)
plt.pie(y)
plt.show()
```

```python
plt.subplot(2, 1, 1)
plt.plot(x, y, 'gH--')
plt.plot(y, x, 'cx-')
plt.subplot(2, 1, 2)
plt.pie(y)
plt.show()
```

## A complex example

```python
x= [ 1,2,3,4,5,7,9,14]
y= [ 1,4,8,6,5,10,9,11]

plt.subplot(2,3,1)
plt.plot(x,y) # Grfico de lineas
plt.subplot(2,3,2)
plt.plot(x,y,'o') # Grfico de lineas con puntos
plt.subplot(2,3,3)
plt.pie(x) # Grfico de torta
plt.subplot(2,3,4)
plt.boxplot(x) # Grfico de cajas
plt.subplot(2,3,5)
plt.bar(x,y) # Grfico de barras
plt.subplot(2,3,6)
plt.hist(x) # Grfico histograma
```
