```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = np.sin(x)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

axes[0].plot(x, y, 'r-.', label='$sin(x)$')
axes[0].set_title('Relacion x - y')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].legend()

axes[1].plot(y, x, 'g:', label='$sin(x)$')
axes[1].set_title('Relacion y - x')
axes[1].set_xlabel('y')
axes[1].set_ylabel('x')
axes[1].legend()

fig.tight_layout()
```

```python
plt.plot(x, y, label='$sin(x$')
plt.title('Relacion x-y')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='upper left', bbox_to_anchor=(1, 0.9))
```
